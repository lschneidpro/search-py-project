```python
from typing import List, Optional, Union

from pydantic import BaseModel, Field, model_validator
from typing_extensions import Self
```

```python
class HierarchicalBucket(BaseModel):
    """
    Bucket representing a hierarchical aggregation.
    """

    key: str = Field(..., description="Key representing a specific value or range.")
    id: Optional[str] = Field(
        None, description="Optional unique identifier for the bucket."
    )
    doc_count: int = Field(..., ge=0, description="Count of documents in this bucket.")
    is_selected: bool = Field(
        False, description="Indicates whether this bucket is selected."
    )
    path: str = Field(..., description="Full hierarchical path of the bucket.")
    type: str = Field(
        ..., description="Type of the bucket, indicating its classification."
    )
    # level: int = Field(..., ge=1, description="Level of this bucket in the hierarchy.")
    children: List["HierarchicalBucket"] = Field(
        default_factory=list,
        description="List of child buckets representing sub-levels in the hierarchy.",
    )
```

```python
def build_hierarchical_tree_with_smart_dummies(data) ->HierarchicalBucket:
    # Step 1: Create a dictionary to store nodes by their path
    node_map = {}

    # Step 2: Process each item and create nodes
    for item in data:
        key_parts = item["key"].split("/")
        node_path = "/".join(key_parts)
        node_type = item["_type"]["buckets"][0]["key"] if item["_type"]["buckets"] else ""
        
        node = HierarchicalBucket(
            key=key_parts[-1],
            doc_count=item["doc_count"],
            path=node_path,
            type=node_type
        )
        node_map[node_path] = node

    # Step 3: Ensure all parents exist by creating smart dummy nodes
    for node_path in list(node_map.keys()):
        parts = node_path.split("/")
        for i in range(1, len(parts)):
            parent_path = "/".join(parts[:i])
            if parent_path not in node_map:
                parent_key = parts[i - 1]
                # Gather children doc_count and type
                children_doc_count = sum(
                    node.doc_count
                    for path, node in node_map.items()
                    if path.startswith(parent_path + "/") and len(path.split("/")) == len(parent_path.split("/")) + 1
                )
                child_types = {
                    node.type
                    for path, node in node_map.items()
                    if path.startswith(parent_path + "/") and len(path.split("/")) == len(parent_path.split("/")) + 1
                }
                # Determine type of dummy node
                dummy_type = child_types.pop() if len(child_types) == 1 else "dummy"

                dummy_node = HierarchicalBucket(
                    key=parent_key,
                    doc_count=children_doc_count,
                    path=parent_path,
                    type=dummy_type
                )
                node_map[parent_path] = dummy_node

    # Step 4: Build the hierarchy by linking children to parents
    root = None
    for node_path, node in node_map.items():
        parent_path = "/".join(node_path.split("/")[:-1])
        if parent_path in node_map:
            node_map[parent_path].children.append(node)
        else:
            if root is None:
                root = node

    return root
```
